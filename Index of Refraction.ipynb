```python
import physplot as phys
import numpy as np
from __future__ import division
import math


incident_angle = np.array([10,20,20,30,40,50,50,60,70])
refracted_angle = np.array([8,15,16,21,28,34,36,39,44])

incident = np.sin(np.radians(incident_angle))
refracted = np.sin(np.radians(refracted_angle))


errangles = np.array([2,2,2,2,2,2,2,2,2])
one = np.sin(np.radians(incident_angle) + np.radians(errangles))
two = incident
errangles = one-two
print(errangles)



#plotfit takes in parameters in this order
#phys.plotfit(x,y,dy,title,'xlabel(units)','ylabel(units)','slope units')  Slope units will not render in LaTeX

phys.plotfit(refracted,incident,errangles,'index of refraction (water)','sin of refracted angle','sin of incident angle','')






```
